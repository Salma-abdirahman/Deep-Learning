## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("Resources/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
#  YOUR CODE GOES HERE
newDataFrame = application_df.drop(columns=['NAME', 'EIN'])
newDataFrame.head()
```

```python
# Determine the number of unique values in each column.
#  YOUR CODE GOES HERE
newDataFrame.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
#  YOUR CODE GOES HERE
ApplicationTypeCounts = newDataFrame['APPLICATION_TYPE'].value_counts()
ApplicationTypeCounts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
#  YOUR CODE GOES HERE
applicationTypesReplacement = ApplicationTypeCounts[ApplicationTypeCounts < 150].index

# Replace in dataframe
for app in applicationTypesReplacement:
    newDataFrame['APPLICATION_TYPE'] = newDataFrame['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
newDataFrame['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
#  YOUR CODE GOES HERE
classificationCount = newDataFrame["CLASSIFICATION"].value_counts()
classificationCount
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
#  YOUR CODE GOES HERE
classificationCount[classificationCount>1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
#  YOUR CODE GOES HERE
classificationsReplacement = classificationCount[classificationCount < 800].index

# Replace in dataframe
for cls in classificationsReplacement:
    newDataFrame['CLASSIFICATION'] = newDataFrame['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
newDataFrame['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
#  YOUR CODE GOES HERE
nemericDummies = pd.get_dummies(newDataFrame)
nemericDummies.head()
```

```python
# Split our preprocessed data into our features and target arrays
#  YOUR CODE GOES HERE
X = nemericDummies.drop('IS_SUCCESSFUL', axis=1)
y = nemericDummies['IS_SUCCESSFUL']

# Split the preprocessed data into a training and testing dataset
#  YOUR CODE GOES HERE
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=44)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE

nn = tf.keras.models.Sequential()

# First hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=80, activation="relu", input_dim=44))

# Second hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=40, activation="relu"))

# Output layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
#  YOUR CODE GOES HERE
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
#  YOUR CODE GOES HERE
fitModel = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
